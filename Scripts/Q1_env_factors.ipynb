```python
#What are the most common environmental factors (weather, lighting, time of day)
#  associated with severe crashes?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#dataset
df = pd.read_csv('../Data/us_accidents_sample_500k_clean.csv')

#print(df.head())

#print(df.info())


#weather vs. Severity == severe
#Weather_Simple & Severity_Label == Severe cleaned in the DataExploration

#filter for only severe
severe = df[df['Severity_Label'] == 'Severe']
#count weather 
weather_count = severe['Weather_Simple'].value_counts().reset_index()
weather_count.columns = ['Weather_Simple', 'Count']
sns.barplot(data=weather_count,
            x='Weather_Simple',
            y='Count',
            palette='husl')
plt.title("Frequency of Severe accidents with different Weathers")
plt.xlabel("Weather Type")
plt.ylabel("Number of Severe Accidents")
plt.show()
```

```python
#lighting vs. Severity
#df.info()

severe['Visibility(mi)'] = pd.to_numeric(severe['Visibility(mi)'], errors='coerce')
#severe['Visibility(mi)'].isna().sum()

severe = severe.dropna(subset=['Visibility(mi)'])

# #use only severe --> used in the above
bins = [0, 1, 3, 5, 10, 20, 50, 100]
labels = ['0-1', '1-3', '3-5', '5-10', '10-20', '20-50', '50-100']

severe['Visibility(mi)'] = pd.cut(severe['Visibility(mi)'], bins=bins, labels=labels)

vis_count = severe['Visibility(mi)'].value_counts().sort_index().reset_index()
vis_count.columns = ['Visibility', 'Count']

sns.barplot(
    data=vis_count,
    x='Visibility',
    y='Count',
    palette='husl'
)
plt.title("Severe Accidents by Visibility")
plt.xlabel("Visibility (mi)")
plt.ylabel("Number of Severe Accidents")
plt.show()
```

```python
#Time of day vs. Severity == Severe
print(df.Sunrise_Sunset.value_counts())

time_count = severe['Sunrise_Sunset'].value_counts().reset_index()
time_count.columns = ['Time_of_Day', 'Count']

sns.barplot(
    data=time_count,
    x='Time_of_Day',
    y='Count',
    palette='bright'
)
plt.title("Severe Accidents by Time of Day")
plt.xlabel("Time of Day")
plt.ylabel("Number of Severe Accidents")
plt.show()
```
